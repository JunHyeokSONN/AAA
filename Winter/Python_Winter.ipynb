```python
# ==== 기본값이 있는 함수 => 강의노트 21쪽 참고

# 4) 임의의 구간에 대해 step 간격으로 합을 구하는 함수  

def f_sum(n, m, step=1):
    hap=0
    for i in range(n,m,step):
        hap=hap+i
    return hap

print("0에서 100미만의 합:", f_sum(0, 100))
print("100+103+106+ … + 199의 합:", f_sum(100, 200, step=3))
print("100+105+110+ … + 199의 합:", f_sum(100, 200, 5))
```

```python
### ============== 다양한 색깔과 크기의 원을 만들어 보기 

import random
import math
import turtle

t=turtle.Turtle()
sc_w = 300
sc_h = 300
turtle.setup(width = sc_w+50, height = sc_h+50)
turtle.screensize(sc_w, sc_h)

t.speed(0)

# 원 20개 표현하기

for i in range(20):
    
    # 임의의 위치 지정
    x=random.randint(-sc_w/2,sc_w/2)
    y=random.randint(-sc_h/2,sc_h/2)

    # 랜덤 색상 지정하기
    r,g,b=random.random(),random.random(),random.random() # rgb 색상 랜덤으로 지정
    t.color(r,g,b) # tt색깔 rgb 지정

    # 랜덤 반지름 지정
    radi = random.randint(0,50)

    # 원 객체 생성
    t.begin_fill()
    t.circle(radi)
    t.end_fill()

    # 임의의 위치로 이동하기
    t.penup()
    t.goto(x,y)
    t.pendown()

turtle.done()
```

```python
# super 사용

class Student :
    def __init__(self , val_name='', val_age=0):
        self.name = val_name
        self.age = val_age

    def get_name(self):
        return self.name
    def get_age(self):
        return self.age

    def set_name(self,name):
        self.name = name
    def set_age(self,age):
        self.age = age

class Girls(Student):

    def __init__(self,name,age,height):
        super().__init__(name, age) # 부모클래스의 객체반환, super는 함수라서 () 해줘야함.
        self.height  = height
        
    def set_height(self, height):
        self.height = height
        
    def get_height(self):
        return self.height

s1 = Girls('lll',10,170)    # __init__ 을 별도로 만들어야함.
print(s1.get_name(),s1.get_age())

s1.set_height(170)
print(s1.get_height())
```

```python
# 캡슐화 - p26
# 변수 이름 앞에 __ 붙여서 데코 하면, 클래스 내부에서만 접근 가능
# __붙여서 사용, 변수이름이 바뀌는거
###### 28p

import math

# 클래스 정의
class Circle:
    def __init__(self, radius=1.0):
        self.__radius = radius # private
    def setRadius (self, r): # private 변수를 설정할 때 사용
        self.__radius = r
    def getRadius(self): # private 변수를 접근할 때 사용
        return self.__radius
    def calcArea(self):
        area = math.pi*self.__radius*self.__radius
        return area
    def calcCircum(self):
        circumference = 2.0*math.pi*self.__radius
        return circumference

#인스턴스 생성
    
#c1=Circle(10)
c1=Circle()
c1.setRadius(10)    # 이런 식으로 줘야함.

print("원의 반지름=", c1.getRadius())
print("원의 넓이=", c1.calcArea())
print("원의 둘레=", c1.calcCircum())

# print("원의 반지름=", c1.__radius)       # 접근 불가
```

```python
################ 3)

class Circle:
    def __init__(self,radius=1):
        self.radius = radius

    def area(self):
        return self.radius * self.radius * 3.14

    def perimeter(self):
        return 2 * self.radius * 3.14

class Cylinder(Circle):

    def __init__(self,radius,height):
        super().__init__(radius)
        self.height = height

    def volume(self):
        return super().area() * self.height

    def sarea(self):
        return super().area()*2 + super().perimeter()*self.height

cy=Cylinder(3,5)
print('원통 체적: %.2f' %cy.volume())
print('원통 표면적: %.2f' %cy.sarea())
```

```python
### 20쪽 마우스 이벤트
### 버튼이벤트 바인딩
# 위젯.bind(event,handler) 왼쪽은 <Button>, 오른쪽은 함수 <= 일반적

from tkinter import *

# function
def btn_fun(event):     # event 변수이름은 아무거나 와도됨.
    txt = ""
    # click event : left / right
    if event.num == 1:
        txt = "left button ("
    elif event.num == 3:
        txt = "right button ("

    # click position
    pos_x = event.x
    pos_y = event.y
    txt = txt + str(pos_x) + ',' + str(pos_y) + ')'

    # label text configure
    la_txt.configure(text=txt)      

# Tk instance
win = Tk()                     # root(base) window
win.title("python window")     # title
win.geometry("400x400")

# bind (base window, button)
win.bind('<Button>',btn_fun)

# label
la_txt = Label(win) # 없어도 configure에서 바꿔줌
la_txt.pack(expand=True, anchor=CENTER)

win.mainloop()
```

```python
##################################### 실습 #####################################
######################### 1)

# 모듈
import random
import turtle
from tkinter import *

# def
def create_poly():
    
    num = int(in_tb.get())
    
    # list
    for i in range(num):
        r,g,b = random.random(),random.random(),random.random()
        t.color(r,g,b)
        angle = random.randint(3,10)        # 삼각형 부터 십각형중 랜덤
        size = random.randint(1,50)         # 한 변 길이 1 ~ 50까지 랜덤

        t.begin_fill()                      # 다각형 그리기
        for j in range(0, angle):
            t.forward(size)
            t.left(360/angle)
        t.end_fill()

        t.penup()                           # 이동
        t.goto(x=random.randint(-250,250), y=random.randint(-250,250))
        t.pendown()
        
# main window
win = Tk()
win.title("Turtle Canvas")

# canvas
drw = Canvas(win, width=500, height=500)

# Label, textbox, button
la_in = Label(win,text='다양한 다각형 표시하기  \n다각형을 표현할 개수 입력 ',font=('',18))
in_tb = Entry(win)
btn_ok = Button(win,text='확인',command=create_poly)

# 위치
drw.pack()
la_in.pack(side=LEFT)
in_tb.pack(side=LEFT)
btn_ok.pack(side=LEFT)

# turtle
t = turtle.RawTurtle(drw)
t.speed(0)

# 마무리
win.mainloop()
```

```python
######################### 1) 변형
  
# 모듈
import random
import turtle
from tkinter import *

# def
def create():
    c_cnt = 0
    num = int(in_tb.get())
    c_cnt = random.randint(0,num)
    p_cnt = num - c_cnt
    # list
    for i in range(c_cnt):
        
        r,g,b = random.random(),random.random(),random.random()
        t.color(r,g,b)
        radi = random.randint(0,50)
        
        t.begin_fill()
        t.circle(radi)
        t.end_fill()
        
        t.penup()
        t.goto(x=random.randint(-250,250), y=random.randint(-250,250))
        t.pendown()
        
    for i in range(p_cnt):
        r,g,b = random.random(),random.random(),random.random()
        t.color(r,g,b)
        angle = random.randint(3,10)        # 삼각형 부터 십각형중 랜덤
        size = random.randint(1,100)     # 한 변 길이 1 ~ 100까지 랜덤

        t.begin_fill()
        for j in range(0,angle):
            t.forward(size)
            t.left(360/angle)
        t.end_fill()

        t.penup()
        t.goto(x=random.randint(-250,250), y=random.randint(-250,250))
        t.pendown()
        
# main window
win = Tk()
win.title("Turtle Canvas")

# canvas
drw = Canvas(win, width=500, height=500)

# Label, textbox, button
la_in = Label(win,text='다양한 원 또는다각형 표시하기\n표현할 도형의 개수 입력 ',font=('',18))
in_tb = Entry(win)
btn_ok = Button(win,text='확인',command=create)

# 위치
drw.pack()
la_in.pack(side=LEFT)
in_tb.pack(side=LEFT)
btn_ok.pack(side=LEFT)

# turtle
t = turtle.RawTurtle(drw)
t.speed(0)

# 마무리
win.mainloop()
```

```python
######################### 2)

# 모듈
import random
from tkinter import *

# 기본 변수
com = random.randint(1,100)
cnt = 1

# def guess
def func_guess():
    
    global cnt
    guess = int(in_tb.get())

    if guess == com:
        text = '빙고!~ 축하합니다 {} 번 만에 맞췄네요~'.format(cnt)
    elif guess > com:
        cnt += 1
        text = '틀렸습니다.\n더 작은 수를 선택하세요.'
    else:
        cnt += 1
        text = '틀렸습니다.\n더 큰 수를 선택하세요.'

    la_btn.configure(text=text)
    in_tb.delete(0,5)
    
# def reset
def func_reset():
    
    global cnt
    global com
    
    cnt = 1
    com = random.randint(1,100)
    la_btn.configure(text='게임을 다시 시작합니다')

# main window
win = Tk()
win.title("컴퓨터가 숨긴 수를 맞추는 게임")
win.geometry("300x200")

# Label, textbox, button
la_in = Label(win,text='컴퓨터가 숨긴수를 맞춰보세요')
in_tb = Entry(win)
btn_ok = Button(win,text='확인',bg='Yellow',command=func_guess)
btn_reset = Button(win,text='게임 다시 시작',bg='Yellow',command=func_reset)
la_btn = Label(win,text='1 ~ 100 사이 컴퓨터가 숨긴 수를 맞춰보세요')

# 위치
la_in.pack(padx=5,pady=5)
in_tb.pack(padx=5,pady=5)
btn_ok.pack(padx=5,pady=5)
btn_reset.pack(padx=5,pady=5)
la_btn.pack(padx=5,pady=5)

# 마무리
win.mainloop()
```
